```python
PG = np.load('./plot_results/PG_RewardCurve.npy')
PG_PPO = np.load('./plot_results/PG_PPO_RewardCurve.npy')
pg_n = 50
pg_ppo_n = 50
pg_step = np.array(range(len(PG_PPO)-pg_n))
pg_ppo_step = np.array(range(len(PG_PPO)-pg_ppo_n))
pg_curve = []
pg_ppo_curve = []
for i in range(len(PG_PPO)-pg_n):
    pg_curve.append(np.mean(PG[i:i+pg_n]))
for i in range(len(PG_PPO)-pg_ppo_n):
    pg_ppo_curve.append(np.mean(PG_PPO[i:i+pg_ppo_n]))
pg_curve = np.array(pg_curve)
pg_ppo_curve = np.array(pg_ppo_curve)
plt.figure(figsize=(8,4))
plt.plot(pg_step, pg_curve, 'b', label = 'PG')
plt.plot(pg_ppo_step, pg_ppo_curve, 'r', label = 'PPO PG')
plt.xlabel('episodes')
plt.ylabel('rewards')
plt.legend()
plt.title('Comparison of PG models')
plt.savefig('PG models')
plt.show()
```

```python
DQN = np.load('./plot_results/resultsP/dqn_save_reward.npy')
gamma_DQN = np.load('./plot_results/resultsP/dqn_gamma79_save_reward.npy')
relu_DQN = np.load('./plot_results/resultsP/dqn_LeakyreluToRelu_save_reward.npy')
h256_DQN = np.load('./plot_results/resultsP/dqn_linearHidden256_save_reward.npy')
m100_DQN = np.load('./plot_results/resultsP/dqn_MemoryCapacity_100_save_reward.npy')
dqn_n = 500

long1 = len(DQN)
long2 = len(gamma_DQN)
long3 = len(relu_DQN)
long4 = len(h256_DQN)
long5 = len(m100_DQN)

# long1 = 1300
# long2 = 1300
# long3 = 1300
# long4 = 1300

dqn_step = np.array(range(long1-dqn_n))
gamma_step = np.array(range(long2-dqn_n))
relu_step = np.array(range(long3-dqn_n))
h256_step = np.array(range(long4-dqn_n))
m100_step = np.array(range(long5-dqn_n))

dqn_curve = []
gamma_dqn_curve = []
relu_dqn_curve = []
h256_dqn_curve = []
m100_dqn_curve = []

for i in range(long1-dqn_n):
    dqn_curve.append(np.mean(DQN[i:i+dqn_n]))
for i in range(long2-dqn_n):
    gamma_dqn_curve.append(np.mean(gamma_DQN[i:i+dqn_n]))
for i in range(long3-dqn_n):
    relu_dqn_curve.append(np.mean(relu_DQN[i:i+dqn_n]))
for i in range(long4-dqn_n):
    h256_dqn_curve.append(np.mean(h256_DQN[i:i+dqn_n]))
for i in range(long5-dqn_n):
    m100_dqn_curve.append(np.mean(m100_DQN[i:i+dqn_n]))

dqn_curve = np.array(dqn_curve)
gamma_dqn_curve = np.array(gamma_dqn_curve)
relu_dqn_curve = np.array(relu_dqn_curve)
h256_dqn_curve = np.array(h256_dqn_curve)
m100_dqn_curve = np.array(m100_dqn_curve)
plt.figure(figsize=(10,6))
plt.plot(dqn_step, dqn_curve, label='origin DQN')
plt.plot(gamma_step, gamma_dqn_curve, 'r', label='gamma=0.79')
plt.plot(relu_step, relu_dqn_curve, 'g', label='leaky ReLU => ReLU')
plt.plot(h256_step, h256_dqn_curve, 'y', label='hidden size=256')
plt.plot(m100_step, m100_dqn_curve, 'b', label='memory capacity=100')
plt.xlabel('episodes')
plt.ylabel('rewards')
plt.title('Comparison of DQN parameters')
plt.legend()
plt.savefig('DQN parameters')
plt.show()
```
